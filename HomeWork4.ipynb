```python
data = df['waterfront'].value_counts()
data.index = ['no', 'yes']
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.pie(data, autopct='%1.1f%%')

plt.title('Доля домов с видом на набережную')
plt.legend(data.index)
```

у большиства домов отсутсвует вид на набережную

```python
data_f = df['floors'].value_counts()
f_val = data_f.values
f_ind = data_f.index
plt.bar(f_ind,f_val, width=0.3)
plt.title('распределение этажей домов')
plt.xlabel('Этажи')
plt.ylabel('Количество домов')
```

большая часть домов одно и двух-этажные

```python
data_с = df['condition'].value_counts()
с_val = data_с.values
с_ind = data_с.index
plt.bar(с_ind,с_val, width=0.3)

plt.title('распределение состояния домов')
plt.xlabel('Состояние')
plt.ylabel('Кол-во')
```

большинство домов в среднем состоянии. Плохих и очень плохих совсем мало.
